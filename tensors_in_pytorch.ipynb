## Reduction operation

```python
e = torch.randint(size=(2,3),low=0,high=10,dtype=torch.float32)
e
```

```python
# sum
torch.sum(e)
```

```python
# sum along columns
torch.sum(e,dim=0)
```

```python
# sum along row
torch.sum(e, dim=1)
```

```python
# mean
e.mean()
```

```python
# mean column wise
e.mean(dim=0)
```

```python
# mean row wise
e.mean(dim=1)
```

```python
# median
e.median()
```

```python
# max and min
e.max(), e.min()
```

```python
# product
e.prod()
```

```python
# standard deviation
e.std()
```

```python
# variance
e.var()
```

```python
# argmax - give max element position
e.argmax()
```

## Matrix operation

```python
f = torch.randint(size=(2,3),low=0,high=10)
g = torch.randint(size=(3,2),low=0,high=10)
f,g
```

```python
# matrix multipilcaton
f.matmul(g)
```

```python
# dot product of vector
vector1 = torch.tensor([1,2])
vector2 = torch.tensor([3,4])
vector1.dot(vector2)
```

```python
# treanspos
f,f.transpose(0,1)
```

```python
# determinant
h = torch.randint(size=(3,3),low=0,high=10,dtype=torch.float32)
h.det()
```

```python
# inverse
h.inverse()
```
