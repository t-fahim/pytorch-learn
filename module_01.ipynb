```python
# creating random tensor of size(3,4) using torch.rand()
r_tensor = torch.rand(size=(3,4))
r_tensor, r_tensor.ndim, r_tensor.dtype
```

```python
# creating random tensor of size (120,240,3)
image = torch.rand(size=(120,240,3))
image.shape, image.ndim
```

```python
# create a tensor of all zero
zeros = torch.zeros(size=(3,4))
zeros,zeros.dtype
```

```python
# create a tensor of all ones
ones = torch.ones(size=(3,4))
ones, ones.dtype
```

```python
# creating a range of values like 0 to 10
# torch.arange(start, end, step)
zero_to_ten = torch.arange(start=0,end=11,step=1)
zero_to_ten
```

```python
# can create a tensor of zeros similar to another tensor
ten_zero = torch.zeros_like(input=zero_to_ten)
ten_zero
```

## Tensor datatypes

```python
# float 32 tensor (defaul)
float_32_tensor = torch.tensor([3.0,6.0,9.0],
                               dtype=None,
                               device=None,
                               requires_grad=False)
float_32_tensor, float_32_tensor.dtype
```

```python
# float 16 tensor
float_16_tensor = torch.tensor([3,5,7],dtype=torch.float16)
float_16_tensor, float_16_tensor.dtype
```
